```python
# Load Data
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Creating a dataset with duplicates,missing values, and outliers...
data = {
    "id": [101, 102, 103, 104, 105],
    "age": [25, 32, 55, 42, 43],
    "name": ["Alice","Bob","Charlie","David","Eve"],
    "date": ["2023-01-01","2023-02-15","2023-03-20","2023-04-10","2023-05-05"],
    "category": ["HR","IT","Finance","Marketing","HR"],
    "salary": [50000,60000,120000,80000,90000]
}

df = pd.DataFrame(data)
df
```

```python
# Measure of Central Tendancy...
print(df['salary'].mean())
print(df['salary'].median())
print(df['category'].mode())
```

```python
# Measures the spread...
df['salary'].std(), df['salary'].var()  # Variance and Standard Deviation
```

```python
# Inter quartile range
Q1 = df['salary'].quantile(0.25)
Q3 = df['salary'].quantile(0.75)
IQR = Q3 - Q1
print(Q1, Q3, IQR)
```

```python
# Histogram ---> To understand data distribution...
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8,5))
sns.histplot(df['salary'],bins=10, kde=True)
plt.title("Histogram of salary")
plt.show()
```

```python
# Boxplot ---> To detect outliers...
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8,5))
sns.boxplot(x = df['salary'])
plt.title("Boxplot of salary")
plt.show()
```

```python
# Scatter Plot ---> To understand data distribution...
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8,5))
sns.scatterplot(x = df['salary'], y = df['age'])
plt.title("Scatter Plot")
plt.show()
```

```python
# Correlation Heatmap ---> Detecting Relationships...
import matplotlib.pyplot as plt
import seaborn as sns

# select only numerical columns
df_numeric = df.select_dtypes(include=['number'])

# Compute correlation matrix
# corr_matrix = df_numeric.corr()

plt.figure(figsize=(10,6))
sns.heatmap(df_numeric.corr(), annot=True, cmap="coolwarm", linewidths=0.5)
plt.title("Correlation Heatmap")
plt.show()
```
